## Google Books Data

```python
import requests
```

```python
url = 'https://www.googleapis.com/books/v1/volumes'
#get the response from the endpoint URL
response = requests.get(url)
```

```python
params = {'q' : 'Machine Learning'}
```

```python
response =requests.get(url, params=params)
response.json()['items']
```

```python
#get the list of the books in the response 
books = response.json()['items']
total = 0
for book in books:
    title = book['volumeInfo']['title']
    num_pages = book['volumeInfo']['pageCount']
    total += num_pages
    if not title:
        print('No title found')

#avrage number of pages for ML books 
avrage = total/len(books)
print(avrage)
```
